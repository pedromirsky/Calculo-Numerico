```python
import sympy as sy
from sympy import sin,exp
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd



sy.init_printing()
```

```python
import sympy as sy
t = sy.Symbol('t')
p = sy.Function('p')
mu = sy.Symbol('mu') # definir

ed_bacterias = sy.Eq(p(t).diff(),mu*p(t))
#ed_bacterias
sy.dsolve(ed_bacterias,p(t))

CI={p(0):200}
sy.dsolve(ed_bacterias,p(t), ics=CI)
```
