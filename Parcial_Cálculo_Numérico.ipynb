## B) Utilizando el método de la bisección:

### Defino la función para el método Bisección:

```python
#x = symbols('x')

def biseccion(a,b,f,iteracionMax = 10000, tol = 0):
  print(f"Hallando raíz dentro del intervalo ( {a} , {b})")
  #x = symbols('x')

  valores_a = []
  valores_b = []
  valores_c = []
  valores_fc = []
  valores_e = []

  datos = {
    'a' : valores_a,
    'b' : valores_b,
    'c' : valores_c,
    'fc': valores_fc,
    'e' : valores_e
    }

  fa = f.subs(x,a)
  fb = f.subs(x,b)
  c = (a + b) / 2
  raizHallada = False
  contador = 0    
  
  while raizHallada == False: # ver de sacar el ==False
      c = (a + b) / 2
      contador += 1
      

      valores_a.append(a)
      valores_b.append(b)
      valores_c.append(c)

      if  len(valores_e) > 0 : # Para que no me ponga primer valor del array de e = 0. (q despues me arruina el tema de <= tol... hace q me corte en la primera iteracion)
        e = abs(c-valores_c[contador-2])
      else:
        e = abs(c)

      valores_e.append(e)

      # print(fa * fb)
      if fa * fb < 0:
        fc = f.subs(x,c)
        
        valores_fc.append(fc)
        

     
        # if fc >= -0.5 and fc <= 0.5: # ACA CREO QUE HAY Q COMPARAR FC PREVIO CON ACTUAL! VER CALSES GRABADAS 
        # 
        # print(e)
        if e <= tol:
        #if fc == 0: # el programa termina solo cuando se da esta condicion
          print("iteraciones: ")
          print(contador)
          print("raiz: ") 
          print(c)
          raizHallada = True

        else:  
          # print(fc)
          if fa * fc < 0:
            b = c
          else:
            a = c
       
      if contador >= iteracionMax: # corta cuando cumplen las iteraciones maximas
          print("iteraciones se cortaron en el limite maximo: ")
          print(contador)
          print("raiz mas cercana: ") 
          print(c)
          raizHallada = True

  df = pd.DataFrame(datos)
  df = df.astype(float) # si no no me plotea fc, no lo toma como numeric.

  #df['a'].plot()
  #df['b'].plot() 
  #df['c'].plot() 
  #df['fc'].plot()  
  #df['e'].plot()  

  #return df.head(10) 
  #return df['fc']

```

## C) Utilizando el método de punto fijo:

### Defino la función para el método del Punto Fijo:

```python
from sympy import *
import pandas as pd

# Trabajo con g(x) = f(x) + x

def puntoFijo (g, p0 = 0, iteracionMax = 10000, tol = 0):
    x = symbols('x')
    valores_p = []
    valores_p0 = []
    valores_e = []

    datos = {
        "p" : valores_p,
        "p0": valores_p0,
        'e' : valores_e
        }


    for i in range(iteracionMax):
    
      p = g.subs(x,p0)
      valores_p.append(p)
      valores_p0.append(p0)
      valores_e.append(abs(p-p0))

      print(p)

      if abs(p-p0) <= tol:
        print(p)
        print("iteraciones: " + str(i+1))
        break;
      else: 
        p0 = p

    df = pd.DataFrame(datos)
    df=df.astype(float) 
    df.head(10)
    df['e'].plot() 


  
```

### Una forma de saber si este método convergerá hacia una de las raíces, es según la condición de convergencia para dicho método, que afirma lo siguiente:

### Sea g Є C[a, b] y g(x) Є [a, b] " x Є [a, b]. Si $ g’(x) en (a, b) y $ una constante positiva k < 1 tal que |g’(x)| ≤ k " x Є (a b), entonces para cualquier número ro en [a, b], la sucesión definida por rn = g(pn-1) converge al único punto fijo p en [a, b].

### Dicho de otra forma, el módulo de la derivada de g(x) debe ser menor a 1 (pendiente de y=x) para que haya convergencia, siendo esta una condición necesaria pero no suficiente.

```python
x = symbols('x')

# vectorEFG= [tolerancia, iteraciones, raíz]
def biseccionError(a,b,f,iteracionMax = 10000, tol = 0):
  #print(f"Tol: ( {tol} )")
  #x = symbols('x')

  valores_a = []
  valores_b = []
  valores_c = []
  valores_fc = []
  valores_e = []

  vectorEFG = [] # este vector contiene La Tolerancia establecida, la raíz hallada, la cant de iteraciones para hallar esa raiz 
  vectorEFG.append(tol)

  datos = {
    'a' : valores_a,
    'b' : valores_b,
    'c' : valores_c,
    'fc': valores_fc,
    'e' : valores_e
    }


  fa = f.subs(x,a)
  fb = f.subs(x,b)
  c = (a + b) / 2
  raizHallada = False
  contador = 0    
  
  while raizHallada == False: # ver de sacar el ==False
      c = (a + b) / 2
      contador += 1
      

      valores_a.append(a)
      valores_b.append(b)
      valores_c.append(c)

      if  len(valores_e) > 0 : # Para que no me ponga primer valor del array de e = 0. (q despues me arruina el tema de <= tol... hace q me corte en la primera iteracion)
        e = abs(c-valores_c[contador-2])
      else:
        e = abs(c)

      valores_e.append(e)

      # print(fa * fb)
      if fa * fb < 0:
        fc = f.subs(x,c)
        
        valores_fc.append(fc)
        

     
        # if fc >= -0.5 and fc <= 0.5: # ACA CREO QUE HAY Q COMPARAR FC PREVIO CON ACTUAL! VER CALSES GRABADAS 
        # 
        # print(e)
        if e <= tol:
        #if fc == 0: # el programa termina solo cuando se da esta condicion
          #print("iteraciones: ")
          #print(contador)
          vectorEFG.append(contador)

          #print("raiz: ") 
          #print(c)
          vectorEFG.append(c)
          raizHallada = True

        else:  
          # print(fc)
          if fa * fc < 0:
            b = c
          else:
            a = c
       
      if contador >= iteracionMax: # corta cuando cumplen las iteraciones maximas
          #print("iteraciones se cortaron en el limite maximo: ")
          #print(contador)
          vectorEFG.append(contador)
          #print("raiz mas cercana: ") 
          #print(c)
          vectorEFG.append(c)
          raizHallada = True

  df = pd.DataFrame(datos)
  df = df.astype(float) # si no no me plotea fc, no lo toma como numeric.

  #df['a'].plot()
  #df['b'].plot() 
  #df['c'].plot() 
  #df['fc'].plot()  
  #return df['e'].plot()  
  return valores_e
  
  
  #return vectorEFG 

  #return df.head(10) 
  #return df['fc']

  # Para crear un df de tolerancia, iteraciones, raiz, a partir de a,b y f
  # en función de las tolerancias que brinda el enunciado
def biseccionDf(a,b,f):
      tolerancias = []
      iteraciones = []
      raiz = []

      

      toleranciasEnunciado = [ 10**(-2), 10**(-5), 10**(-7), 10**(-12)]

      for i in toleranciasEnunciado:
        vectorEFG = biseccionTolIterRaiz(a,b, f, tol =i)
        tolerancias.append(vectorEFG[0])
        iteraciones.append(vectorEFG[1])
        raiz.append(vectorEFG[2])

      BiseccionEFG = {
          'Tolerancia' : tolerancias,
          'Iteraciones' : iteraciones,
          'Raiz' : raiz
          }
      df = pd.DataFrame(BiseccionEFG)
      df = df.astype(float) 

      return df
```
