```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1.0, 3.0, num=50)
f = np.exp(-x)-x
f1 = np.exp(-x)
f2 = x

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('m√©todo punto fijo')

ax1.plot(x, f, 'o-')
ax1.set_ylabel('f')
ax1.axhline(y=0, color='k', linestyle='-')

ax2.plot(x, f1, '.-', label="e-x")
ax2.plot(x, f2, '.-', label="x")
ax2.set_xlabel('x')
ax2.set_ylabel('f')

plt.legend()
plt.show()
```
