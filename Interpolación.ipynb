```python
import sympy as sp
x = sp.sympify('x')
f = 1/x

x0 = 2
x1 = 2.75
x2 = 4

y0 = f.subs(x,x0)
y1 = f.subs(x,x1)
y2 = f.subs(x,x2)
```

```python
L0 =  ( (x-x1)  /  (x0-x1) ) *  ( (x-x2)  /  (x0-x2) )
```

```python
L1 =  ( (x-x0)  /  (x1-x0) ) * ( (x-x2)  /  (x1-x2) )
```

```python
L2 =  ( (x-x0)  /  (x2-x0) ) * ( (x-x1)  /  (x2-x1) )
```

```python

pn = L0 * y0 + L1 * y1 + L2 * y2

pn = sp.simplify(pn)
pn
```

```python
# primero guardo los valores obtenidos en vectores 
# Lk = [L0, L1, L2]

x = sp.sympify('x')
f = 1/x

x0 = 2
x1 = 2.75
x2 = 4

y0 = f.subs(x,x0)
y1 = f.subs(x,x1)
y2 = f.subs(x,x2)

Y = [y0, y1, y2]
X = [x0, x1, x2]


Pn = 0



for k in range (len(Y)):
  productoria = 1

  for i in range (len(Y)):

    if i != k:

      factor = ( ( x-X[i] ) / (X[k] - X[i]) )


      productoria = productoria * factor 
      
      #termino =  (Y[k] * ( ( x-X[i] ) / (X[k] - X[i]) ))
      #Pn = Pn + termino

  termino =  Y[k] * productoria
  Pn = Pn + termino
  
Pn = sp.simplify(Pn)
Pn

  





 
```
